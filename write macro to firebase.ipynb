```python
def reverse(lst):
    return [ele for ele in reversed(lst)]
```

```python
def returndata(given_url):
    url = given_url
    r = requests.get(url)
    data = r.json()
    print(data)
    return data['data']
```

```python
def returndate(oldlist,newlist):
    newlist=[]
    newlist_str=str(newlist)
    for i in range (len(oldlist)):
        newlist.append(oldlist[i]['date'])
    return newlist



def returnvalue(oldlist,newlist):
    newlist=[]
    newlist_str=str(newlist)
    for i in range (len(oldlist)):
#         print(oldlist[i]['date'])
        newlist.append(oldlist[i]['value'])
#         print(newlist)
    return newlist
```
